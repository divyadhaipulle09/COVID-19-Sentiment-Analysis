```python
import pandas as pd
import re
import string
from nltk import word_tokenize, ngrams
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer

# Define the clean_text function to preprocess the text
def clean_text(text):
    # Remove URLs
    text = re.sub(r"http\S+", "", text)
    # Remove usernames
    text = re.sub(r"@[^\s]+", "", text)
    # Remove hashtags
    text = re.sub(r"#", "", text)
    # Remove punctuation, lowercase, and tokenize
    text = text.translate(str.maketrans('', '', string.punctuation))
    text = text.lower()
    tokens = word_tokenize(text)
    # Remove stop words
    stopwords_remove = stopwords.words('english') + ['thats', 'weve', 'hes', 'theres', 'ive', 'im','to'
                                                     'will', 'can', 'cant', 'dont', 'youve', 'us', 
                                                     'youre', 'youll', 'theyre', 'whats', 'didnt']
    tokens = [token for token in tokens if token not in stopwords_remove]
    # Join tokens back into a string
    text_cleaned = ' '.join(tokens)
    return text_cleaned

# Define the frequent_terms function to extract unigrams, bigrams, or trigrams from text
def frequent_terms(text, ngram=1, min_df=5):
    # Preprocess the text
    text_cleaned = [clean_text(t) for t in text]
    # Vectorize the text
    vectorizer = CountVectorizer(ngram_range=(ngram, ngram), min_df=min_df)
    X = vectorizer.fit_transform(text_cleaned)
    # Get the feature names and counts
    feature_names = vectorizer.get_feature_names()
    counts = X.sum(axis=0).tolist()[0]
    # Convert to a DataFrame
    df = pd.DataFrame({'term': feature_names, 'count': counts})
    df = df.sort_values(by='count', ascending=False)
    return df


```

```python
from nltk.corpus import stopwords
import pandas as pd
import numpy as np
import re
import string
import seaborn as sns
import matplotlib.pyplot as plt

# Load and preprocess data
data = tweets
data = data.dropna(subset=['text'])
data['text'] = data['text'].apply(lambda x: re.sub('[%s]' % re.escape(string.punctuation), '', x))
stopwords_remove = stopwords.words('english') + ['thats', 'weve', 'hes', 'theres', 'ive', 'im','to',
                                                 'will', 'can', 'cant', 'dont', 'youve', 'us', 
                                                 'youre', 'youll', 'theyre', 'whats', 'didnt']
data['text'] = data['text'].apply(lambda x: ' '.join([word for word in x.split() if word.lower() not in stopwords_remove]))

# Train LDA model
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation

vectorizer = CountVectorizer(max_features=10000)
X = vectorizer.fit_transform(data['text'])

lda_model = LatentDirichletAllocation(n_components=5, max_iter=50, random_state=0)
lda_model.fit(X)

# Print topics and top words in each topic
top_words = [[vectorizer.get_feature_names()[i] for i in topic.argsort()[:-11:-1]] for topic in lda_model.components_]
topic_names = {i: ' '.join(words) for i, words in enumerate(top_words)}
for i, words in enumerate(top_words):
    print(f'Topic {i}:')
    print(words)

# Assign topic to each document
topics = lda_model.transform(X)
data['topic'] = np.argmax(topics, axis=1)

# Plot the distribution of topics
sns.countplot(x='topic', data=data)
plt.show()
```
